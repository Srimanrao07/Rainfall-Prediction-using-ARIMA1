```python
# extract out the time-series
Rainfall_ts = Rainfall['ANNUAL']
plt.figure(figsize=(10, 5))
plt.plot(Rainfall['ANNUAL'])
plt.title("trend or seasonality")
plt.xlabel('YEAR')
plt.ylabel('ANNUAL RAINFALL IN cms')
```

```python
# remove wrinkles from our time series using moving average.
# moving average of different time periods i.e. 4,6,8, and 12 months
fig, axes = plt.subplots(2, 2, sharey=False, sharex=False)
fig.set_figwidth(14)
fig.set_figheight(8)
axes[0][0].plot(Rainfall_ts.index, Rainfall_ts, label='Original')
axes[0][0].plot(Rainfall_ts.index, Rainfall_ts.rolling(window=4).mean(), label='4-Months Rolling Mean')
axes[0][0].set_xlabel("Years")
axes[0][0].set_ylabel("ANNUAL RAINFALL IN cms")
axes[0][0].set_title("4-Months Moving Average")
axes[0][0].legend(loc='best')

axes[0][1].plot(Rainfall_ts.index, Rainfall_ts, label='Original')
axes[0][1].plot(Rainfall_ts.index, Rainfall_ts.rolling(window=12).mean(), label='12-Months Rolling Mean')
axes[0][1].set_xlabel("Years")
axes[0][1].set_ylabel("ANNUAL RAINFALL IN cms")
axes[0][1].set_title("12-Months Moving Average")
axes[0][1].legend(loc='best')

axes[1][0].plot(Rainfall_ts.index, Rainfall_ts, label='Original')
axes[1][0].plot(Rainfall_ts.index, Rainfall_ts.rolling(window=16).mean(), label='16-Months Rolling Mean')
axes[1][0].set_xlabel("Years")
axes[1][0].set_ylabel("ANNUAL RAINFALL IN cms")
axes[1][0].set_title("16-Months Moving Average")
axes[1][0].legend(loc='best')

axes[1][1].plot(Rainfall_ts.index, Rainfall_ts, label='Original')
axes[1][1].plot(Rainfall_ts.index, Rainfall_ts.rolling(window=60).mean(), label='60-Months Rolling Mean')
axes[1][1].set_xlabel("Years")
axes[1][1].set_ylabel("ANNUAL RAINFALL IN cms")
axes[1][1].set_title("60-Months Moving Average")
axes[1][1].legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
'''# 60-month moving average produces a wrinkle free curve.
# There is expected monthly-seasonal effect in our data.'''
# plot the rolling mean and standard deviation on window of 600 months.
# Determing rolling statistics
rolmean = Rainfall_ts.rolling(window=60).mean()
rolstd = Rainfall_ts.rolling(window=60).std()
#Plot rolling statistics:
orig = plt.plot(Rainfall_ts, label='Original')
mean = plt.plot(rolmean, label='Rolling Mean')
std = plt.plot(rolstd, label = 'Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show(block=False)
```

```python
# Dicky Fuller Test on the timeseries
# run the Dicky Fuller Test on the timeseries and
# verify the null hypothesis that the TS is non-stationary.
# Perform Dickey-Fuller test:
from statsmodels.tsa.stattools import adfuller

print('Results of Dickey-Fuller Test:')
dftest = adfuller(Rainfall_ts, autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#lags Used', 'Number of Observations Used'])

for key, value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
import itertools
import statsmodels.api as sm
import statsmodels.tsa.api as smt
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
plt.style.use('bmh')
from io import StringIO
import requests
```

```python
Rainfall_ts_log_diff = Rainfall_ts.diff(periods=1) # same as ts_log_diff = ts_log - ts_log.shift(periods=1)
Rainfall_ts_log_diff.dropna(inplace=True)
fig, axes = plt.subplots(1, 2, sharey=False, sharex=False)
fig.set_figwidth(12)
fig.set_figheight(4)
smt.graphics.plot_acf(Rainfall_ts_log_diff, lags=30, ax=axes[0], alpha=0.5)
smt.graphics.plot_pacf(Rainfall_ts_log_diff, lags=30, ax=axes[1], alpha=0.5)
plt.tight_layout()
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf

lag_acf = acf(Rainfall_ts_log_diff, nlags=20)
lag_pacf = pacf(Rainfall_ts_log_diff, nlags=20, method='ols')

plt.figure(figsize=(10,8))

#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(Rainfall_ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(Rainfall_ts_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')

#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(Rainfall_ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(Rainfall_ts_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

```python
# Step 8: Forecast sales using the best fit ARIMA model
n_steps = 36
pred_uc_99 = best_results.get_forecast(steps=36, alpha=0.01) # alpha=0.01 signifies 99% confidence interval
pred_uc_95 = best_results.get_forecast(steps=36, alpha=0.05) # alpha=0.05 95% CI
```

```python
# Get confidence intervals 95% & 99% of the forecasts
pred_ci_99 = pred_uc_99.conf_int()
pred_ci_95 = pred_uc_95.conf_int()
n_steps = 36
idx = pd.date_range(Rainfall_ts.index[-1], periods=n_steps, freq='MS')
fc_95 = pd.DataFrame(np.column_stack([np.power(10, pred_uc_95.predicted_mean), np.power(10, pred_ci_95)]),
index=idx, columns=['forecast', 'lower_ci_95', 'upper_ci_95'])
fc_99 = pd.DataFrame(np.column_stack([np.power(10, pred_ci_99)]),
index=idx, columns=['lower_ci_99', 'upper_ci_99'])
fc_all = fc_95.combine_first(fc_99)
fc_all = fc_all[['forecast', 'lower_ci_95', 'upper_ci_95', 'lower_ci_99', 'upper_ci_99']] # just reordering columns
fc_all.head()
```
